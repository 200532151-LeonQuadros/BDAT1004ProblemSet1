```python
#-----------------Question 1-----------------------

#Step 1. Import the necessary libraries
import pandas as pd

#Step 2. Import the dataset from this address.
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'

users = pd.read_csv(url, sep='|')

#Step 3. Assign it to a variable called users
users = pd.read_csv(url, sep='|')

```

```python
#Step 4. Discover what is the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
mean_age_per_occupation

```

```python
#Step 5. Discover the Male ratio per occupation and sort it from the most to the least
male_ratio_per_occupation = users[users['gender'] == 'M'].groupby('occupation')['gender'].count() / users.groupby('occupation')['gender'].count()
male_ratio_per_occupation.sort_values(ascending=False)

```

```python
#Step 6. For each occupation, calculate the minimum and maximum ages
min_max_ages_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
min_max_ages_per_occupation
```

```python
#Step 7. For each combination of occupation and gender, calculate the mean age
mean_age_per_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()

mean_age_per_occupation_gender
```

```python
#Step 8. For each occupation present the percentage of women and men
percentage_per_occupation = users.groupby(['occupation', 'gender'])['gender'].count() / users.groupby('occupation')['gender'].count()
percentage_per_occupation
```

```python
#-----------------Question 8-----------------------

import pandas as pd
import matplotlib.pyplot as plt

# Collect and organize data
data = {
    'Actor': ['Tom Cruise', 'Jackie Chan', 'Jason Statham', 'Bruce Willis', 'Sylvester Stallone'],
    'Kills': [65, 107, 188, 132, 278]
}
df = pd.DataFrame(data)

# Create horizontal bar chart
plt.barh(df['Actor'], df['Kills'])

# Sort actors by number of kills
df = df.sort_values(by='Kills')

# Label each bar with actor's name
plt.yticks(df['Actor'])

# Add axes labels
plt.xlabel('Number of Kills')
plt.ylabel('Actor')

# Show bar chart
plt.show()
```

```python
#-----------------Question 9-----------------------

import matplotlib.pyplot as plt

# Set the number of Roman Emperors that were assassinated
num_assassinated = 69

# Set the total number of Roman Emperors
total_emperors = 420

# Calculate the fraction of assassinated emperors
fraction_assassinated = num_assassinated / total_emperors

# Create the pie chart
plt.pie([fraction_assassinated, 1 - fraction_assassinated], labels=["Assassinated", "Not assassinated"], 
        autopct="%1.1f%%")

# Make sure the pie chart is an even circle
plt.axis("equal")

# Show the percentage breakdown of the categories
plt.legend(title="Percentage breakdown")

# Show the pie chart
plt.show()


```
