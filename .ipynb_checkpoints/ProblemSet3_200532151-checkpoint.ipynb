```python
#-----------------Question 1-----------------------

#Step 1. Import the necessary libraries
import pandas as pd

#Step 2. Import the dataset from this address.
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'

users = pd.read_csv(url, sep='|')

#Step 3. Assign it to a variable called users
users = pd.read_csv(url, sep='|')

```

```python
#Step 4. Discover what is the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
mean_age_per_occupation

```

```python
#Step 5. Discover the Male ratio per occupation and sort it from the most to the least
male_ratio_per_occupation = users[users['gender'] == 'M'].groupby('occupation')['gender'].count() / users.groupby('occupation')['gender'].count()
male_ratio_per_occupation.sort_values(ascending=False)

```

```python
#Step 6. For each occupation, calculate the minimum and maximum ages
min_max_ages_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
min_max_ages_per_occupation
```

```python
#Step 7. For each combination of occupation and gender, calculate the mean age
mean_age_per_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()

mean_age_per_occupation_gender
```

```python
#Step 8. For each occupation present the percentage of women and men
percentage_per_occupation = users.groupby(['occupation', 'gender'])['gender'].count() / users.groupby('occupation')['gender'].count()
percentage_per_occupation
```

```python
#-----------------Question 2-----------------------

#Step 1. Import the necessary libraries
import pandas as pd

#Step 2. Import the dataset from this address
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv'
euro12 = pd.read_csv(url)

#Step 3. Assign it to a variable called euro12
euro12 = pd.read_csv(url)




```

```python
#Step 6. What is the number of columns in the dataset?
num_columns = len(euro12.columns)
print(num_columns)

```

```python
#Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
discipline
```

```python


#Step 8. Sort the teams by Red Cards, then to Yellow Cards
discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)

```

```python
#Step 9. Calculate the mean Yellow Cards given per Team
yellow_cards_mean = discipline['Yellow Cards'].mean()
print(yellow_cards_mean)

```

```python
#Step 10. Filter teams that scored more than 6 goals
high_scoring_teams = euro12[euro12['Goals'] > 6]
high_scoring_teams
```

```python
#Step 11. Select the teams that start with G
teams_starting_with_g = euro12[euro12['Team'].str.startswith('G')]
teams_starting_with_g
```

```python

#Step 12. Select the first 7 columns
first_7_columns = euro12.iloc[:, :7]
first_7_columns
```

```python

#Step 13. Select all columns except the last 3
all_columns_except_last_3 = euro12.iloc[:, :-3]

all_columns_except_last_3
```

```python
#Step 14. Present only the Shooting Accuracy from England, Italy and Russia
shooting_accuracy = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])]['Shooting Accuracy']
shooting_accuracy
```

```python
#-----------------Question 8-----------------------

import matplotlib.pyplot as plt
import pandas as pd

# Read data from CSV file
df = pd.read_csv('https://raw.githubusercontent.com/200532151-LeonQuadros/BDAT1004ProblemSet_200532151/main/actor_kill_counts.csv')
df.columns = ['Actor', 'Kill Count']

# Create horizontal bar chart
plt.barh(df['Actor'], df['Kill Count'], tick_label=df['Actor'])

# Label axes
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Dealiest Actors in Hollywood')

# Display plot
plt.show()

```

```python
#-----------------Question 9-----------------------

import matplotlib.pyplot as plt
import pandas as pd


# Read in data from csv file
data = pd.read_csv('https://raw.githubusercontent.com/200532151-LeonQuadros/BDAT1004ProblemSet_200532151/main/roman-emperor-reigns.csv')

# Count number of assassinated emperors
assassinated = data['Cause_of_Death'].value_counts()['Assassinated']

# Count total number of emperors
total = data.shape[0]

# Calculate fraction of emperors assassinated
fraction_assassinated = assassinated/total

# Create list of labels
labels = ['Assassinated', 'Other']

# Create list of sizes
sizes = [fraction_assassinated, 1-fraction_assassinated]

# Create figure
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Make sure pie chart is an even circle
ax1.axis('equal')

# Add title
plt.title('Fraction of Roman Emperors Assassinated')

# Show plot
plt.show()


```

```python
#-----------------Question 10-----------------------

import matplotlib.pyplot as plt
import pandas as pd

# Read data from CSV file
df = pd.read_csv('https://raw.githubusercontent.com/200532151-LeonQuadros/BDAT1004ProblemSet_200532151/main/arcade-revenue-vs-cs-doctorates.csv')
df.columns=['Year','Total Arcade Revenue (billions)','Computer Science Doctorates Awarded (US)']

# Create scatter plot
plt.scatter(df['Total Arcade Revenue (billions)'], df['Computer Science Doctorates Awarded (US)'], c=df['Year'])

# Label axes
plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Computer Science Doctorates Awarded (US)')

plt.title('Revenue of Arcade vs Computer Science Doctorates Awarded (US)')

# Display plot
plt.show()

```
